# ðŸ¦· Running a Custom GGUF Model on Ollama (Local Guide)

This guide shows you how to take a `.gguf` model (e.g. exported from Unsloth) and run it locally using [Ollama](https://ollama.com/).

---

## âœ… Step-by-Step to Create & Run Your Model

### 1. Create a local directory to hold your model

```bash
mkdir -p ~/ollama-models/doctor-dental
cd ~/ollama-models/doctor-dental
```

### 2. Download your GGUF model file

From Hugging Face (replace URL if needed):

```bash
wget https://huggingface.co/BirdieByte1024/YOUR_MODEL/resolve/main/unsloth.F16.gguf -O doctor-dental.gguf
```

Or manually download via browser and move it:

```bash
mv ~/Downloads/unsloth.F16.gguf ~/ollama-models/doctor-dental/doctor-dental.gguf
```

---

### 3. Create a `Modelfile` in the same folder

Run:

```bash
nano Modelfile
```

Paste this content:

```Dockerfile
FROM ./doctor-dental.gguf

TEMPLATE """<|im_start|>system
You are a helpful and knowledgeable medical assistant.<|im_end|>
<|im_start|>user
{{ .Prompt }}<|im_end|>
<|im_start|>assistant
"""

PARAMETER stop "<|im_end|>"
```

Then save:
- Press `CTRL + O`, then `Enter`
- Press `CTRL + X` to exit

---

### 4. Build the model with Ollama

```bash
ollama create doctor-dental -f Modelfile
```

This will register your model internally with Ollama.

---

### 5. Run your model!

```bash
ollama run doctor-dental
```
